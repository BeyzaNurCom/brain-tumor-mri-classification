```python
import tensorflow as tf

model_path = "/content/drive/MyDrive/deeplearningproject/models/final_baseline_cnn_model.keras"
model = tf.keras.models.load_model(model_path)
```

```python
CLASSES = ["glioma", "meningioma", "notumor", "pituitary"]
```

```python
import numpy as np
import gradio as gr
from PIL import Image

IMG_SIZE = (224, 224)

def predict(image: Image.Image):

    image = image.convert("RGB").resize(IMG_SIZE)


    x = np.array(image).astype("float32") / 255.0


    x = np.expand_dims(x, axis=0)


    probs = model.predict(x, verbose=0)[0]
    top_idx = int(np.argmax(probs))
    top_label = CLASSES[top_idx] if top_idx < len(CLASSES) else str(top_idx)


    return {CLASSES[i]: float(probs[i]) for i in range(len(CLASSES))}, f"Tahmin: {top_label} (p={probs[top_idx]:.4f})"
```

```python
iface = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="pil", label="Girdi Görseli (RGB)"),
    outputs=[
        gr.Label(num_top_classes=4, label="Sınıf Olasılıkları"),
        gr.Textbox(label="Özet")
    ],
    title="Baseline CNN (.keras) - Drive Model"
)

iface.launch()
```
